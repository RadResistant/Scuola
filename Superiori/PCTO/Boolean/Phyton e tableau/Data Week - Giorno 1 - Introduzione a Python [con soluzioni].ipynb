### Strutture di Controllo

Le strutture di controllo sono degli strumenti fondamentali da avere nel proprio arsenale. Permettono al programmatore di interagire con gli oggetti e le variabili del programma e far prendere una strada piuttosto che un'altra in base a delle condizioni prestabilite. 

Le strutture di controllo di base sono: 
- **sequenziali**: questa è la situazione di default che abbiamo già incontrato, una serie di righe ri codice che si susseguono in un programma;
- **condizionali**: rappresentano dei criteri di *selezione*, vengono usati per eseguire ona o più righe di codice **se** una certa condizione è soddisfatta;
- **cicliche**: aggiungono un elemento *iterativo*, il cui scopo è quello di ripetere una certa sequenza di azioni **per (for)** un certo numero di volte o **fintanto che (while)** una certa condizione viene soddisfatta. 

<img src="https://drive.google.com/uc?export=view&id=1HFIxzouzbDuD_0vq1lqaJIbXLiWLQT-J" width="800">

#### Condizionali

Le strutture condizionali permettono di controllare il flusso del codice sulla base di **condizioni booleane** (vero - falso) che vengono specificate dal programmatore. 

Per inizializzare una struttura condizionale utilizzeremo l'espressione `if`; qui sotto vediamo un esempio in pseudocodice: 

```
if <condizione booleana>:
    <riga (o righe) di codice>
```

Vediamo un esempio reale che usa la sintassi di Python:

````python
# attenzione all'indentazione!
if 10 > 1: 
    print("Corretto!")
````

Cose da notare su questa struttura: 

- inizia sempre con la parola `if` 
- la parola `if` è sempre seguita dall'espressione booleana che dev'essere valutata
- alla fine dell'espressione booleana vanno sempre messi i `:`
- il codice che dev'essere eseguito nel caso in cui l'espressione booleana sia vera (il corpo della struttura `if`) dev'essere **indentato** 

Vediamo un altro esempio dove andiamo prima a definire una variabile `a` e poi la confrontiamo con un altro valore e, se una certa condizione è soddisfatta, stampiamo a schermo una risposta: 

````python
a = 10
if a < 20: 
    print("a è minore di 20")
````

E se `a` è maggiore di 20, che cosa succede? 

````python
a = 30
if a < 20: 
    print("a è minore di 20")
````

In questo caso non succede nulla perché non abbiamo specificato un percorso **alternativo (else)**. L'espressione `else` permette di dire al programma che cosa deve fare nel caso in cui la condizione booleana non sia rispettata (cioè sia falsa): 

````python
if a < 20: 
    print("a è minore di 20")
else: 
    print("a è maggiore di 20")
````

Cose da ricordare: 

- la parola `else` deve essere allo stesso livello (di indentatura) della parola `if`
- dopo la parola `else` non c'è bisogno di specificare alcuna condizione booleana 
- i `:` devono sempre seguire la parola `else` 
- il codice da eseguire nel caso in cui l'espressione booelana non sia rispettata deve essere **indentato** 

È anche possibile specificare più di un'alternativa; ad esempio, potremmo specificare diversi outup a seconda che `a` sia:

- minore di 10
- tra 10 e 19 (inclusi)
- tra 20 e 29 (inclusi)
- maggiore o uguale a 30

Per fare ciò introduciamo l'espressione `elif`, la quale ci permette di specificare più condizioni:

````python
a = 15
if a < 10: 
    print("a è minore di 10")
elif 10 <= a <= 19: 
    print("a è tra 10 e 19 (inclusi)")
elif 20 <= a <= 29: 
    print("a è tra 20 e 29 (inclusi)")
else: 
    print("a è maggiore o uguale a 30")
````

Cose da tenere a mente: 
- la parola `elif` deve essere allo stesso livello d'indentazione della parola `if` 
- la parola `elif` è sempre seguita dall'espressione booleana che dev'essere valutata
- alla fine dell'espressione booleana vanno sempre messi i `:`
- il codice che dev'essere eseguito nel caso in cui l'espressione booleana sia vera dev'essere **indentato** 

Con questi tre nuovi strumenti, `if`, `elif` ed `else` possiamo costruire dei flussi condizionali di grande complessità sulla base di logiche predeterminate dal programmatore. 

#### Cicli

I cicli, strutture molto comuni in tutti i linguaggi di programmazione, permettono di ripetere una sequenza di righe di codice **per (for)** un certo numero di volte o **fintanto che (while)** una certa condizione rimane soddisfatta. 

Come sarà chiaro dalle righe precendenti, esistono due categorie principali di strutture iterative: i cicli for e i cicli while. 

**Ciclo for**

Il ciclo `for` viene usato per iterare su una sequenza di elementi (come una lista, un dizionario, una stringa o anche un range di numeri) doce il numero di iterazioni fatte (e che rimangono da fare) viene controllato da un contatore (o indice); contestualmente, ad ogni iterazione viene eseguito un pezzo di codice. 

Per inizializzare un ciclo for utilizzeremo l'espressione `for`; qui sotto vediamo un esempio in pseudocodice: 

```
for <contatore> in <sequenza>:
    <riga (o righe) di codice>
```

Vediamo un esempio in Python: 

````python
# attenzione all'indentazione!
for lettera in "parola": 
    print(lettera)
````

Cose da ricordare: 

- inizia sempre con l'espressione `for`
- va sempre specificato un contatore (in questo caso è stato chiamato `lettera`, ma è possibile assegnargli qualsiasi altro nome)
- è necessario avere una sequenza sulla quale iterare il contatore (in questo esempio la sequenza è una stringa, quindi una sequenza di lettere, ma anche liste, tuples, e set sono ottimi candidati)
- contatore e sequenza sono sempre seguiti da un `:`
- la riga o le righe di codice che devono essere eseguite in ciascuna iterazione devono essere **indentate** 

Dietro le quinte il contatore `lettera` và a iterare sugli elementi della stringa "parola" (cioè su ciascuna lettera) e ogni volta che incontra un nuovo elemento esegue il codice (cioè stampa a schermo la lettera).

Come detto più sopra, il `contatore` è solo un alias, quindi possiamo chiamarlo come vogliamo, può essere qualsiasi parola o può anche essere formato da una sola lettera: 

````python
for i in "stringa": 
    print(i)
````

In molti contesti è necessario ripetere un'operazione per un certo numero di volte; in questi casi è opportuno usare la funzione `range()`:

````python
for i in range(1, 4): 
    print("Beetlejuice!")
````

La funzione `range()` richiede tre parametri in input: 

```
range(start, stop, step)
```

dove: 
- `start` è l'inizio (incluso) del range numerico (parametro opzionale, default = 0)
- `stop` è la fine (non incluso) del range numerico 
- `step` indica di l'ampiezza dell'incremento (opzionale, default = 1). 

Come detto sopra, il punto finale `stop` non viene incluso nella sequenza:

````python
for num in range(0, 6):
    print(num)
````

Come già annunciato, qualsiasi oggetto iterabile può essere incluso come `iteratore` in un ciclo for. Vediamo un esempio con una lista:

````python
lista = [1,2,3,4,5]
print(lista)
````

````python
for el in lista: 
    print(el, el**2)
````

**Ciclo While** 

Un ciclo while permette di eseguire una o più righe di codice fintanto che una certa condizione booleana rimane verificata; solitamente viene usato un contatore che, alla fine di ciascuna iterazione, viene incrementato (o ridotto) sulla base di una certa logica decisa dal programmatore. Una volta che la condizione booleana smette di essere verificata, il programma esce dal ciclo e si ferma. 

Per inizializzare un ciclo while utilizzeremo l'espressione `while`; qui sotto vediamo un esempio in pseudocodice: 

```
while <condizione booleana>:
    <riga (o righe) di codice>
```

Vediamo un esempio con la sintassi di Python: 


````python
a = 1
while a < 6:
    print(a)
    a = a + 1
````

È anche possibile specificare una condizione `else`, nel caso in cui vogliamo far fare qualcosa al programma (come stampare un messaggio) non appena la condizione booleana smetta di essere verificata: 

````python
a = 1
while a < 6: 
    print(a)
    a = a + 1
else: 
    print('stop:', a, "è maggiore di 5")
````

> **Attenzione**: nel caso in cui la condizione booleana sia sempre verificata (eg: se avessimo scritto: a > 0 nell'esempio precedente), il programma non uscirà mai dal ciclo e andrà avanti all'infinito (o finché non verrà arrestato dal programmatore).




Nel corpo di entrambi i cicli `for` e `while`, è possibile inserire delle strutture condizionali (e viceversa); ci sono diverse ragioni per le quali potresti volerlo fare:
- se vuoi fermare il ciclo preventivamente, puoi combinare un `if` con un `break` 
- per saltare l'iterazione corrente, puoi combinare l'`if` con un `continue`

````python
# break 
a = 1
while a < 6: 
    print(a)
    a = a + 1
    if a == 3: 
        break
````

````python
# continue 
a = 0
while a < 6: 
    a = a + 1
    if a == 3: 
        continue
    print(a)
````

### Funzioni ⚡︎

Spesso ci si ritrova a **scrivere e riscrivere sempre le stesse righe** di codice, magari con piccole variazioni di alcuni parametri. In questi casi, invece di continuare a fare copia e incolla dal codice precedente, può essere una buona idea **scrivere una funzione**. 

Le funzioni sono dei **blocchi di codice** ai quali viene assegnato un nome (il nome della funzione) e che possono prendere in input dei parametri (ma non è obbligatorio). 

Quindi, in sostanza, una funzione prende dei valori in input, li passa al blocco di codice che è stato scritto (quello che ci eravamo stancati di scrivere e riscrivere), e produce un output come risultato finale. 

<img src="https://drive.google.com/uc?export=view&id=1L82Af-NassPOom_xKYp0doE_b3ygX7qB" width="400">

In Python è possibile **definire una funzione** usando l'espressione `def`: 

````python
def la_mia_prima_funzione(): 
    print("Ciao, sono l'output della funzione!")
````

Nella cella qui sopra abbiamo **definito** una funzione e, come avrai notato, non ha prodotto alcun output. Questo perché è stata definita ma non è stata utilizzata (in gergo si dice che non è stata **chiamata**). 

Per chiamare una funzione, basta scrivere il suo nome seguito da delle parentesi (e, se ce ne sono, specificare i parametri di input):

````python
la_mia_prima_funzione()
````

Come dicevamo, una funzione può prendere dei **parametri in input**, in altre parole possiamo decidere di passare delle informazioni aggiuntive all'interno delle parentesi della funzione di modo che possano essere usati da questa: 

````python
def saluto(nome): 
    print("Ciao", nome)
````

````python
saluto(nome="Booleaner")
````

Naturalmente una funzione può anche restituire dei valori numerici come output, scriviamo una funzione che prende due numeri in input e restituisce il loro prodotto: 

````python
def prodotto(a, b): 
    a * b
````

````python
prodotto(4, 5)
````

Nella cella precedente abbiamo chiamato la funzione ma non è stato restituito alcun risultato. Questo perché per poter restituire un risultato, la parte risultante dev'essere preceduta dall'espressione `return`: 

````python
def prodotto(a, b): 
    return a * b
````

````python
prodotto(4, 5)
````

## Esercizio pratico ⚡︎

Proviamo a creare un **generatore di password** automatico. Per generare delle buone password dovremo poter specificare il numero di caratteri e dovrà avere le seguenti caratteristiche: 

*  lettere minuscole
*  lettere maiuscole
*  valori numerici
*  caratteri speciali

Per prima cosa carichiamo le librerie di cui avremo bisogno e poi andiamo a definire delle stringhe contenenti le quattro tipologie di caratteri che abbiamo definito qui sopra:

````python
import string
import random

lower = string.ascii_lowercase
upper = string.ascii_uppercase
numbers = '0123456789'
special = '!$%?#@^_&'
````

A questo punto uniamo le quattro variabili appena create in un'unica stringa usando l'operatore `+` di concatenazione:

````python
concat = lower + upper + numbers + special
````

La funzione `random.sample(population, k)` permette di estrarre in maniera casuale `k` elementi da una sequenza (come la sequenza di caratteri in una stringa) la quale viene passata alla funzione tramite il primo parametro `population`:

````python
random.sample(population=lower, k=4)
````

L'operatore `.join()` permette di concatenare diversi elementi contenuti in una lista o una stringa e separarli con un carattere a scelta. Ad esempio, prendiamo la lista contenente le prime tre lettere `['a','b','c']` e proviamo a concatenarle usando il carattere `'_'` come separatore: 

````python
'_'.join(['a','b','c'])
````

Se invece che usare il carattere `'_'` usassimo una stringa vuota `''`, le tre lettere verrebbero concatenate senza spazi o altri caratteri tra loro: 

````python
''.join(['a','b','c'])
````

Possiamo quindi usare questo metodo per concatenare i caratteri che sono stati estratti in maniera casuale usando la funzione `random.sample()`: 

````python
"".join(random.sample(concat, 12))
````

Proviamo ora a racchiudere i passi appena visti in una funzione che possa essere riutilizzata su richiesta: 

````python
def generate_password(len = 8): 
  lower = string.ascii_lowercase
  upper = string.ascii_uppercase
  numbers = '0123456789'
  special = '!$%?#@^_&'

  concat = lower + upper + numbers + special

  return "".join(random.sample(concat, len))
````

````python
generate_password(12)
````
